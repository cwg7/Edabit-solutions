## 2)

```python
def cpp_txt(lst):
	newLst = list(lst[:-1])
	word = ''
	for i in newLst:
		word += i
	return word
```

```python
if __name__=='__main__':
	print(cpp_txt(["H", "e", "l", "l", "o", "!", "\0"]))
```

## 3)

```python
def find_none(lst):
    x = None
    for i in lst:
        if x in lst:
            return lst.index(x)
        else:
            return -1
```

```python
if __name__=='__main__':
	print(find_none([3, 5, 7, None, 7]))
```

## 4)

```python
def sum_minimums(lst):
	return sum(min(x) for x in lst)
```

```python
if __name__=='__main__':
	print(sum_minimums([[1, 2, 3, 4, 5],
                   [5,6,7,8,9],
                   [20,21,34,56,100]]))
```

## 5)

```python
def card_hide(card):
	firstPart = card[0:-4]
	length = len(firstPart)
	lastFour = card[-4:]
	return '*' * length + lastFour
```

```python
if __name__=='__main__':
	print(card_hide("1234123456785678"))
```

```python
def card_hide(card):
	return '*'*len(card[:-4])+card[-4:]
```
