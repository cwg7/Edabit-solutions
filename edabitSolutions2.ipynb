## 2)

```python
def cpp_txt(lst):
	newLst = list(lst[:-1])
	word = ''
	for i in newLst:
		word += i
	return word
```

```python
if __name__=='__main__':
	print(cpp_txt(["H", "e", "l", "l", "o", "!", "\0"]))
```

## 3)

```python
def find_none(lst):
    x = None
    for i in lst:
        if x in lst:
            return lst.index(x)
        else:
            return -1
```

```python
if __name__=='__main__':
	print(find_none([3, 5, 7, None, 7]))
```

## 4)

```python
def sum_minimums(lst):
	return sum(min(x) for x in lst)
```

```python
if __name__=='__main__':
	print(sum_minimums([[1, 2, 3, 4, 5],
                   [5,6,7,8,9],
                   [20,21,34,56,100]]))
```

## 6)

```python
def card_hide(card):
	firstPart = card[0:-4]
	length = len(firstPart)
	lastFour = card[-4:]
	return '*' * length + lastFour
```

```python
if __name__=='__main__':
	print(card_hide("1234123456785678"))
```

```python
## cleaner solution:
```

```python
def card_hide(card):
	return '*'*len(card[:-4])+card[-4:]
```

 ## 9)

```python
def sortedSquared(array):
    s = sorted(array)
    newArr = []
    for i in s:
        newArr.append(i**2)
    return sorted(newArr)
```

```python
if __name__=='__main__':
	print(sortedSquared([1,2,3,5,6,7,8,9]))
```

## 10)

```python
def remove_numbers(string):
    return "".join(i for i in string if i not in '0123456789')
```

```python
if __name__=='__main__':
	print(remove_numbers('12and2rew984'))
```

```python
def is_first_superior(lst1, lst2):
	return lst1 > lst2
```

```python
if __name__=='__main__':
	print(is_first_superior([1,2,3,4], [1,2,4,4]))
```

```python
if __name__=='__main__':
	print(is_first_superior([1,2,5,4], [1,2,4,4]))
```

## 12)

```python
def assign_person_to_job(names, jobs):
	res = {names[i]: jobs[i] for i in range(len(names))}
	return res
```

```python
if __name__=='__main__':
	print(assign_person_to_job(['Annie', 'Steven', 'Lisa', 'Osman'], ['Teacher', 'Engineer', 'Doctor', 'Cashier']))
```
