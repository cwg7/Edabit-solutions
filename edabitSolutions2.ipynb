```python
def print_list(n):
	return list(range(1, n+1, 1))
```

```python
if __name__=='__main__':
	print(print_list(10))
```

## 2)

```python
def cpp_txt(lst):
	newLst = list(lst[:-1])
	word = ''
	for i in newLst:
		word += i
	return word
```

```python
if __name__=='__main__':
	print(cpp_txt(["H", "e", "l", "l", "o", "!", "\0"]))
```
