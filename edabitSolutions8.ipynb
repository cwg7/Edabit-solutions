```python
import math

def get_distance(a, b):
	side1 =  b['x'] - a['x']
	side2 = b['y'] - a['y']
	result = math.sqrt(side1**2 + side2**2)
	return round(result, 3)
    
```

```python
if __name__=='__main__':
    print(get_distance({"x": -2, "y": 1}, {"x": 4, "y": 3}))
```

```python
def halflife_calculator(mass, hlife, n):
	originalN = n
	while n > 0:
		mass/= 2
		n-=1
	finalMass = round(mass, 3)
	return [finalMass, hlife * originalN]
```

```python
if __name__=='__main__':
	print(halflife_calculator(1600, 6, 3))
```

```python
def sum_of_two(a, b, v):
	for i in a:
		for j in b:
			if i + j == v:
				return True
	return False
```

```python
if __name__=='__main__':
	print(sum_of_two([1, 2], [4, 5, 6], 8))
```

```python
## second solution:
```

```python
def sum_of_two(a, b, v):
	return any(i+j == v for i in a for j in b)
```

```python
if __name__=='__main__':
	print(sum_of_two([1, 2], [4, 5, 6], 8))
```

```python
import string

lowers = string.ascii_lowercase
uppers = string.ascii_uppercase
allLetters = lowers + uppers
numbers = '1234567890'

def count_all(txt):
	letterCount = 0
	numberCount = 0
	for i in txt:
		if i in allLetters:
			letterCount +=1
		if i in numbers:
			numberCount +=1
	return {"LETTERS": letterCount, "DIGITS": numberCount}
```

```python
if __name__=='__main__':
	print(count_all("H3ll0 Wor1d"))
```

```python
## second solution:
```

```python
def count_all(txt):
    return {
    'LETTERS': sum(1 for x in txt if x.isalpha()),
    'DIGITS': sum(1 for x in txt if x.isnumeric()),
  }
```

```python
if __name__=='__main__':
	print(count_all("H3ll0 Wor1d"))
```

```python
def get_only_evens(nums):
	evenIndex = []
	evenNums = []
	finalLst = []
	for i in nums:
		if nums.index(i) % 2 == 0:
			evenIndex.append(i)
	for i in nums:
		if i % 2 == 0:
			evenNums.append(i)
	for i in evenNums:
		if i in evenIndex:
			finalLst.append(i)
	tempSet = set(finalLst)
	return list(tempSet)
```

```python
if __name__=='__main__':
	print(get_only_evens([1,3,2,6,4,8]))
```

```python
def get_only_evens(nums):
	newLst = []
	index = 0
	for i in nums:
		if i%2==0 and index%2==0:
			newLst.append(i)
			index+=1
		else:
			index+=1
			continue
	return newLst
```

```python
if __name__=='__main__':
	print(get_only_evens([49, 41, 30, 44, 26, 0, 5]))
```

```python
def one_odd_one_even(n):
	toStr = str(n)
	last = int(toStr[-1])
	first = int(toStr[0])
	return last%2==0 and first%2!=0 or last%2!=0 and first%2==0
```

```python
if __name__=='__main__':
	print(one_odd_one_even(12))
```

```python
def divisible_by_b(a, b):
	tempVar = a
	newLst = []
	while len(newLst) == 0:
		if tempVar % b == 0:
			newLst.append(tempVar)
			return tempVar
		else:
			tempVar+=1
```

```python
if __name__=='__main__':
	print(divisible_by_b(17,8))
```

```python
import math
def line_length(dot1, dot2):
	distance = math.sqrt( ((dot1[0]-dot2[0])**2)+((dot1[1]-dot2[1])**2) )
	return round(distance,2)
```

```python
def find_it(items, name):
	name1 = name.title()
	if len(items) == 0:
		return name1 + ' is here!'
	elif name in items:
		return name1 + ' is gone...'
	else:
		return name1 + ' is here!'
```

```python
if __name__=='__main__':
	print(find_it(items = {
  "tv": 30,
  "timmy": 20,
  "stereo": 50,
}, name='timmy'))
```
