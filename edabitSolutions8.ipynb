```python
import math

def get_distance(a, b):
	side1 =  b['x'] - a['x']
	side2 = b['y'] - a['y']
	result = math.sqrt(side1**2 + side2**2)
	return round(result, 3)
    
```

```python
if __name__=='__main__':
    print(get_distance({"x": -2, "y": 1}, {"x": 4, "y": 3}))
```

```python
def halloween(dt):
	lst = [dt.split('/')]
	return 'Bonfire toffee' if lst[0][1] == '10' and lst[0][2] =='31' else 'toffee'
```

```python
if __name__=='__main__':
    print(halloween("2013/10/31"))
```

## 5)

```python
def halflife_calculator(mass, hlife, n):
	originalN = n
	while n > 0:
		mass/= 2
		n-=1
	finalMass = round(mass, 3)
	return [finalMass, hlife * originalN]
```

```python
if __name__=='__main__':
	print(halflife_calculator(1600, 6, 3))
```

## 6)

```python
def sum_of_two(a, b, v):
	for i in a:
		for j in b:
			if i + j == v:
				return True
	return False
```

```python
if __name__=='__main__':
	print(sum_of_two([1, 2], [4, 5, 6], 8))
```

```python
## second solution:
```

```python
def sum_of_two(a, b, v):
	return any(i+j == v for i in a for j in b)
```

```python
if __name__=='__main__':
	print(sum_of_two([1, 2], [4, 5, 6], 8))
```
