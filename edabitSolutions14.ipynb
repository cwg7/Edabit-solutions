```python
def cap_space(txt):
	return "".join([i if i.islower() else ' ' + i.lower() for i in txt])
```

```python
if __name__=='__main__':
    print(cap_space('helloWorld'))
```

```python
def dice_game(lst):
	result = 0
	for i in lst:
		if i[0]==i[1]:
			return 0
		else:
			result += i[0] + i[1]
	return result
```

```python
if __name__=='__main__':
    print(dice_game([(1, 2), (3, 4), (5, 6)]))
```
