```python
def cap_space(txt):
	return "".join([i if i.islower() else ' ' + i.lower() for i in txt])
```

```python
if __name__=='__main__':
    print(cap_space('helloWorld'))
```

## 2)

```python
def dice_game(lst):
	result = 0
	for i in lst:
		if i[0]==i[1]:
			return 0
		else:
			result += i[0] + i[1]
	return result
```

```python
if __name__=='__main__':
    print(dice_game([(1, 2), (3, 4), (5, 6)]))
```

## 3)

```python
def is_central(txt):
	return txt == txt[::-1]
```

```python
if __name__=='__main__':
    print(is_central("  #  "))
```

## 4)

```python
def match_last_item(lst):
	newLst = list(str(i) for i in lst)
	newLst = newLst[:-1]
	return ''.join(newLst) == lst[-1]
```

```python
if __name__=='__main__':
    print(match_last_item(["rsq", "6hi", "g", "rsq6hig"]))
```
