```python
def cap_space(txt):
	return "".join([i if i.islower() else ' ' + i.lower() for i in txt])
```

```python
if __name__=='__main__':
    print(cap_space('helloWorld'))
```

## 2)

```python
def dice_game(lst):
	result = 0
	for i in lst:
		if i[0]==i[1]:
			return 0
		else:
			result += i[0] + i[1]
	return result
```

```python
if __name__=='__main__':
    print(dice_game([(1, 2), (3, 4), (5, 6)]))
```

## 4)

```python
def match_last_item(lst):
	newLst = list(str(i) for i in lst)
	newLst = newLst[:-1]
	return ''.join(newLst) == lst[-1]
```

```python
if __name__=='__main__':
    print(match_last_item(["rsq", "6hi", "g", "rsq6hig"]))
```

## 7)

```python
def calc_bundled_temp(n, temp):
	newLst = list(temp.split('*'))
	val = int(newLst[0])
	while n > 0:
		val *= 1.1
		n-=1
	val = round(val, 1)
	val = str(val)
	return val + '*' + ''.join(newLst[1:])
```

```python
if __name__=='__main__':
    print(calc_bundled_temp(1, "2*C"))
```

## 8)

```python
def diving_minigame(lst):
	breath = 10
	for i in lst:
		if i < 0:
			breath -= 2
		else:
			breath = min(breath+4, 10)
		if breath <= 0: return False
	return True
```

```python
if __name__=='__main__':
    print(diving_minigame([-5, -15, -4, 0, 5]))
```

## 9)

```python
def is_alpha(word):
	alphabet = 'abcdefghijklmnopqrstuvwxyz'
	word = word.lower()
	newLst = []
	for i in word:
		if i in alphabet:
			newLst.append(alphabet.index(i)+1)
	return sum(newLst)%2==0
```

```python
if __name__=='__main__':
    print(is_alpha("i'am king"))
```
