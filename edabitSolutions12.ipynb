```python
GUEST_LIST = {
	"Randy": "Germany", 
	"Karla": "France", 
	"Wendy": "Japan", 
	"Norman": "England", 
	"Sam": "Argentina"
}

def greeting(name):
	if name not in GUEST_LIST:
		return "Hi! I'm a guest."
	return "Hi! I'm "+name+", and I'm from "+ GUEST_LIST[name] + "."
```

```python
if __name__=='__main__':
    print(greeting("Randy"))
```

## 3)

```python
def get_missing_letters(s):
	alphabet = 'abcdefghijklmnopqrstuvwxyz'
	result = ''
	for i in alphabet:
		if i not in s:
			result += i
	return result
```

```python
if __name__=='__main__':
    print(get_missing_letters("abcdefgpqrstuvwxyz"))
```

```python
def get_missing_letters(s):
	return ''.join(chr(i) for i in range(97, 123) if chr(i) not in s)
```

```python
if __name__=='__main__':
    print(get_missing_letters("abcdefgpqrstuvwxyz"))
```

## 6)

```python
def count(n):
    if n < 0:
        n = n * -1
    count = 0
    while n > 10:
        n /=10
        count +=1
    return count +1
        
```

```python
if __name__=='__main__':
    print(count(-92563))
```

```python
## second solution:
```

```python
def count(n):
	if abs(n)<10:
		return 1
	return 1+count(abs(n)//10)
```

```python
if __name__=='__main__':
    print(count(-92563))
```

## 7)

```python
def sum_missing_numbers(lst):
	newLst = []
	for i in range(min(lst), max(lst)+1):
		if i not in lst:
			newLst.append(i)
	return sum(newLst)
```

```python
if __name__=='__main__':
    print(sum_missing_numbers([4, 3, 8, 1, 2]))
```

## 9)

```python
def move(word):
	alphabet = 'abcdefghijklmnopqrstuvwxyz'
	newStr = ''
	newIndex = 0
	for i in word:
		newIndex = alphabet.index(i) + 1
		i = alphabet[newIndex]
		newStr += i
	return newStr
```

```python
if __name__=='__main__':
    print(move('hello'))
```

```python
## second solution:
```

```python
def move(word):
	return ''.join(chr(ord(i) + 1) for i in word)
```

```python
if __name__=='__main__':
    print(move('hello'))
```

## 10)

```python
def count_uppercase(lst):
	x = ''.join(lst)
	return sum(1 for i in x if i.isupper())
```

```python
if __name__=='__main__':
    print(count_uppercase(["SOLO", "hello", "Tea", "wHat"]))
```

## 11)

```python
def not_good_math(n, k):
	while k > 0:
		if n % 10 == 0:
			n/=10
			k-=1
		else:
			n -=1
			k-=1
	return int(n)
```

```python
if __name__=='__main__':
    print(not_good_math(540, 5))
```

## 12)

```python
def mineral_formation(cave):
	if sum(cave[0]) > 0 and sum(cave[-1]) == 0:
		return 'stalactites'
	elif sum(cave[0]) == 0 and sum(cave[-1]) > 0:
		return 'stalagmites'
	else:
		return 'both'
```

```python
if __name__=='__main__':
    print(mineral_formation([
  [0, 1, 0, 1],
  [0, 1, 0, 1],
  [0, 0, 0, 1],
  [0, 0, 0, 0]
]))
```

## 15)

```python
def is_repdigit(num):
	toStr = str(num)
	if toStr == '0':
		return True
	for i in toStr:
		if i != toStr[0]:
			return False
	return True
```

```python
if __name__=='__main__':
    print(is_repdigit(66))
```

```python
## second solution:
```

```python
def is_repdigit(num):
    return num>=0 and len(set(str(num)))==1
```

```python
if __name__=='__main__':
    print(is_repdigit(66))
```

## 16)

```python
def countdown(n, txt):
	result = ''
	while n > 0:
		result += str(n)+ '. '
		n-=1
	return result + txt.upper() + '!'
```

```python
if __name__=='__main__':
    print(countdown(10, "Blast Off"))
```
