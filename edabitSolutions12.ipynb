```python
GUEST_LIST = {
	"Randy": "Germany", 
	"Karla": "France", 
	"Wendy": "Japan", 
	"Norman": "England", 
	"Sam": "Argentina"
}

def greeting(name):
	if name not in GUEST_LIST:
		return "Hi! I'm a guest."
	return "Hi! I'm "+name+", and I'm from "+ GUEST_LIST[name] + "."
```

```python
if __name__=='__main__':
    print(greeting("Randy"))
```
