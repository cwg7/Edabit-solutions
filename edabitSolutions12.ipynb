```python
GUEST_LIST = {
	"Randy": "Germany", 
	"Karla": "France", 
	"Wendy": "Japan", 
	"Norman": "England", 
	"Sam": "Argentina"
}

def greeting(name):
	if name not in GUEST_LIST:
		return "Hi! I'm a guest."
	return "Hi! I'm "+name+", and I'm from "+ GUEST_LIST[name] + "."
```

```python
if __name__=='__main__':
    print(greeting("Randy"))
```

## 2)

```python
class Name:
    def __init__(self, fname, lname):
        self.fname = fname.capitalize()
        self.lname = lname.capitalize()
        self.fullname = "{} {}".format(self.fname, self.lname)
        self.initials = "{}.{}".format(self.fname[0], self.lname[0])
```

```python
name1 = Name('jack', 'jones')
```

```python
print(name1.initials)
```

## 3)

```python
def get_missing_letters(s):
	alphabet = 'abcdefghijklmnopqrstuvwxyz'
	result = ''
	for i in alphabet:
		if i not in s:
			result += i
	return result
```

```python
if __name__=='__main__':
    print(get_missing_letters("abcdefgpqrstuvwxyz"))
```
