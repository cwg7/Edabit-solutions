```python
GUEST_LIST = {
	"Randy": "Germany", 
	"Karla": "France", 
	"Wendy": "Japan", 
	"Norman": "England", 
	"Sam": "Argentina"
}

def greeting(name):
	if name not in GUEST_LIST:
		return "Hi! I'm a guest."
	return "Hi! I'm "+name+", and I'm from "+ GUEST_LIST[name] + "."
```

```python
if __name__=='__main__':
    print(greeting("Randy"))
```

```python
class Name:
    def __init__(self, fname, lname):
        self.fname = fname.capitalize()
        self.lname = lname.capitalize()
        self.fullname = "{} {}".format(self.fname, self.lname)
        self.initials = "{}.{}".format(self.fname[0], self.lname[0])
```

```python
name1 = Name('jack', 'jones')
```

```python
print(name1.initials)
```

## 3)

```python
def get_missing_letters(s):
	alphabet = 'abcdefghijklmnopqrstuvwxyz'
	result = ''
	for i in alphabet:
		if i not in s:
			result += i
	return result
```

```python
if __name__=='__main__':
    print(get_missing_letters("abcdefgpqrstuvwxyz"))
```

## 6)

```python
def count(n):
    if n < 0:
        n = n * -1
    count = 0
    while n > 10:
        n /=10
        count +=1
    return count +1
        
```

```python
if __name__=='__main__':
    print(count(-92563))
```

```python
def count(n):
	if abs(n)<10:
		return 1
	return 1+count(abs(n)//10)
```

```python
if __name__=='__main__':
    print(count(-92563))
```

## 7)

```python
def sum_missing_numbers(lst):
	newLst = []
	for i in range(min(lst), max(lst)+1):
		if i not in lst:
			newLst.append(i)
	return sum(newLst)
```

```python
if __name__=='__main__':
    print(sum_missing_numbers([4, 3, 8, 1, 2]))
```
