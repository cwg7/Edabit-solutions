## 1)

```python
def probability(lst, n):
	newLst = sorted(lst)
	lenNewLst = len(newLst)
	count = 0
	for i in newLst:
		if i >= n:
			count +=1
	return round(100 * count / lenNewLst, 1)
```

```python
if __name__=='__main__':
	print(probability([5,1,8,9], 6))
```

```python
## alternative solution
```

```python
def probability(lst, n):
	return round(sum([i >= n for i in lst])/len(lst)*100, 1)
```

```python
if __name__=='__main__':
	print(probability([5,1,8,9], 6))
```

## 2)

```python
def even_odd_string(txt):
    return txt[::2] + ' ' + txt[1::2]   
```

```python
if __name__=='__main__':
	print(even_odd_string("mubashir"))
```

## 3)

```python
def to_number_list(lst):
    newLst = []
    for i in lst:
        if '.' in i:
            newLst.append(float(i))
        else:
            newLst.append(int(i))
    return newLst
            
```

```python
if __name__=='__main__':
	print(to_number_list(["1", "3", "3.6"]))
```

```python
## cleaner solution:
```

```python
def to_float_list(lst):
    return [float(i) for i in lst]
```

```python
if __name__=='__main__':
	print(to_number_list(["1", "3", "3.6"]))
```

## 4)

```python
def is_boiling(temp):
	t = int(temp[:-1])
	return t>=212 if 'F' in temp else t>=100
```

```python
if __name__=='__main__':
	print(is_boiling("214C"))
```

```python
def validate_binary(b):
    return b.count('1') % 2 == 0 and b[-1] == '1' or b[-1] == '0' and b.count('1') % 2 == 0
```

```python
if __name__=='__main__':
	print(validate_binary("00101101"))
```
