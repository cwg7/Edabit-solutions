```python
def probability(lst, n):
	newLst = sorted(lst)
	lenNewLst = len(newLst)
	count = 0
	for i in newLst:
		if i >= n:
			count +=1
	return round(100 * count / lenNewLst, 1)
```

```python
if __name__=='__main__':
	print(probability([5,1,8,9], 6))
```

```python
def probability(lst, n):
	return round(sum([i >= n for i in lst])/len(lst)*100, 1)
```

```python
if __name__=='__main__':
	print(probability([5,1,8,9], 6))
```
