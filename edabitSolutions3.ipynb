## 1)

```python
def find_odd(lst):
	targetNum = 0
	for i in lst:
		if lst.count(i) % 2 != 0:
			targetNum = i
	return targetNum
```

```python
if __name__=='__main__':
	print(find_odd([1,1,2,-2,5,2,4,4,-1,-2,5]))
```

## 2)

```python
def hamming_distance(txt1, txt2):
    count = 0
    for i in range(len(txt1)):
        if txt1[i] != txt2[i]:
            count +=1
    return count
```

```python
if __name__=='__main__':
	print(hamming_distance('abcde', 'abcou'))
```

## 3)

```python
def filter_list(lst):
    return [i for i in lst if isinstance(i, int)]
```

```python
if __name__=='__main__':
	print(filter_list([1,2,'a','b']))
```
