## 1)

```python
def find_odd(lst):
	targetNum = 0
	for i in lst:
		if lst.count(i) % 2 != 0:
			targetNum = i
	return targetNum
```

```python
if __name__=='__main__':
	print(find_odd([1,1,2,-2,5,2,4,4,-1,-2,5]))
```

## 2)

```python
def hamming_distance(txt1, txt2):
    count = 0
    for i in range(len(txt1)):
        if txt1[i] != txt2[i]:
            count +=1
    return count
```

```python
if __name__=='__main__':
	print(hamming_distance('abcde', 'abcou'))
```

## 5)

```python
def count_characters(lst):
	count = 0
	for i in lst:
		count += len(i)
	return count
```

```python
if __name__=='__main__':
	print(count_characters(["###",
                          "###",
                          "###"]))
```

```python
## cleaner solution
```

```python
def count_characters(lst):
    return sum(len(i) for i in lst)
```

```python
if __name__=='__main__':
	print(count_characters(["###",
                          "###",
                          "###"]))
```

## 6)

```python
def move_to_end(lst, el):
	for item in lst:
		if item == el:
			lst.remove(item)
			lst.append(item)
	return lst
```

```python
if __name__=='__main__':
	print(move_to_end([1,3,2,4,4,1], 1))
```

```python
def double_char(txt):
	newWord = ''
	for i in txt:
		newWord += i * 2
	return newWord
```

```python
if __name__=='__main__':
	print(double_char('String'))
```

## 8)

```python
def index_of_caps(word):
    newLst = []
    for i in word:
        if i.isupper() == True:
            newLst.append(word.index(i))
    return newLst
```

```python
if __name__=='__main__':
	print(index_of_caps('eDaBiT'))
```

```python
## another solution
```

```python
def index_of_caps(word):
    return [i for i in range(len(word)) if word[i].isupper()]
```

```python
if __name__=='__main__':
	print(index_of_caps('sUn'))
```

## 9)

```python
def greater_than_one(frac):
    nums = frac.split('/')
    num1 = int(nums[0])
    num2 = int(nums[1])
    return num1 / num2 > 1
```

```python
if __name__=='__main__':
	print(greater_than_one('7/4'))
```

## 11)

```python
def find_difference(a, b):
	if a[0] * a[1] * a[2] > b[0] * b[1] * b[2]:
		return (a[0] * a[1] *a[2]) - (b[0] * b[1] * b[2])
	else:
		return (b[0] * b[1] * b[2])  - (a[0] * a[1] *a[2])
```

```python
if __name__=='__main__':
	print(find_difference([28, 16, 29], [7, 8, 17]))
```

```python
def trimmed_averages(lst):
	lst.remove(min(lst))
	lst.remove(max(lst))
	return mean(lst)
```

```python
def even_last(lst):
	newLst = lst[::2]
	return sum(newLst) * lst[-1] if len(lst) > 0 else 0
```

```python
if __name__=='__main__':
	print(even_last([1,3,3,1,10]))
```

## 15)

```python
def age_difference(ages):
	newLst = sorted(ages, reverse = True)
	difference = newLst[0] - newLst[1]
	if difference == 1:
		return str(difference) + ' year'
	elif difference == 0:
		return 'No age difference between spouses'
	else:
		return str(difference) + ' years'
```

```python
if __name__=='__main__':
	print(age_difference([29, 1, 6, 8, 28]))
```
