```python
import math
def eval_factorial(lst):
	newLst = []
	tot = 0
	for i in lst:
		if '!' in i:
			i = i.replace('!','')
			newLst.append(int(i))
	for i in newLst:
		tot += math.factorial(i)
	return tot
```

```python
if __name__=='__main__':
    print(eval_factorial(["5!", "5!", "10!"]))
```

```python
from math import factorial

def eval_factorial(lst):
	return sum(factorial(int(i.replace('!',''))) for i in lst)
```

```python
if __name__=='__main__':
    print(eval_factorial(["5!", "5!", "10!"]))
```

## 3)

```python
def is_circle_collision(c1, c2):
    r1, x1, y1 = c1
    r2, x2, y2 = c2
    d = ((x1-x2)**2 + (y1-y2)**2)**0.5
    return d <= r1 + r2
```

```python
if __name__=='__main__':
    print(is_circle_collision([10, 0, 0], [10, 10, 10]))
```

```python
def reverse_odd(txt):
	newLst = list(txt.split(' '))
	revWords = []
	for i in newLst:
		if len(i)%2!=0:
			revWords.append(i[::-1])
		else:
			revWords.append(i)
	return ' '.join(revWords)
```

```python
if __name__=='__main__':
    print(reverse_odd("One two three four"))
```

```python
def reverse_odd(txt):
	return ' '.join(i[::-1] if len(i)%2 else i for i in txt.split())
```

```python
if __name__=='__main__':
    print(reverse_odd("One two three four"))
```
