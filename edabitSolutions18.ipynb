```python
import math
def eval_factorial(lst):
	newLst = []
	tot = 0
	for i in lst:
		if '!' in i:
			i = i.replace('!','')
			newLst.append(int(i))
	for i in newLst:
		tot += math.factorial(i)
	return tot
```

```python
if __name__=='__main__':
    print(eval_factorial(["5!", "5!", "10!"]))
```

```python
if __name__=='__main__':
    print(eval_factorial(["5!", "5!", "10!"]))
```

```python
def is_circle_collision(c1, c2):
    r1, x1, y1 = c1
    r2, x2, y2 = c2
    d = ((x1-x2)**2 + (y1-y2)**2)**0.5
    return d <= r1 + r2
```

```python
if __name__=='__main__':
    print(is_circle_collision([10, 0, 0], [10, 10, 10]))
```

## 4)

```python
def reverse_odd(txt):
	newLst = list(txt.split(' '))
	revWords = []
	for i in newLst:
		if len(i)%2!=0:
			revWords.append(i[::-1])
		else:
			revWords.append(i)
	return ' '.join(revWords)
```

```python
if __name__=='__main__':
    print(reverse_odd("One two three four"))
```

```python
## second solution:
```

```python
def reverse_odd(txt):
	return ' '.join(i[::-1] if len(i)%2 else i for i in txt.split())
```

```python
if __name__=='__main__':
    print(reverse_odd("One two three four"))
```

## 6)

```python
def intersection_union(lst1, lst2):
	i = list(set(lst1).intersection(set(lst2)))
	s = list(set(lst1).union(set(lst2)))
	return [sorted(i), sorted(s)]
```

```python
if __name__=='__main__':
    print(intersection_union([1, 2, 3, 4, 4], [4, 5, 9]))
```

## 7)

```python
def duplicate_nums(nums):
	result = []
	for i in nums:
		if nums.count(i)>1:
			result.append(i)
	return sorted(list(set(result))) if len(result)>0 else None
```

```python
if __name__=='__main__':
    print(duplicate_nums([81, 72, 43, 72, 81, 99, 99, 100, 12, 54]))
```

```python
def lcm(a, b):
	searching = True
	x=1
	while searching:
		if x%a==0 and x%b==0:
			return x
		else:
			x+=1
```

```python
if __name__=='__main__':
    print(lcm(4, 6))
```

```python
def batting_avg(lst):
	hits = []
	atbats = []
	for i in lst:
		hits.append((i[0]))
		atbats.append((i[1]))
	x= str(round(sum(hits)/sum(atbats),3))[1:]
	if float(x)*1000%10==0:
		return str(round(sum(hits)/sum(atbats),3))[1:] + '0'
	else:
		return x
```

```python
if __name__=='__main__':
    print(batting_avg([[2, 5], [2, 3], [0, 3], [1, 5], [2, 4]]))
```

```python
def batting_avg(lst):
	a, b = map(sum, zip(*lst))
	return '{:.3f}'.format(a/b)[1:]
```

```python
if __name__=='__main__':
    print(batting_avg([[2, 5], [2, 3], [0, 3], [1, 5], [2, 4]]))
```

```python
def join_digits(n):
	result = ''
	for i in range(1,n+1):
		if i>=10:
			result += str(i)[0]+ '-' + str(i)[1]+ '-'
		else:
			result += str(i) + '-'
	result = result[:-1]
	return result
```

```python
if __name__=='__main__':
    print(join_digits(11))
```

```python
## second solution:
```

```python
def join_digits(n):
	return '-'.join(j for i in range(1, n+1) for j in str(i))
```

```python
if __name__=='__main__':
    print(join_digits(11))
```

```python
def make_sandwich(ingredients, flavor):
    sandwich = []
    for i in ingredients:
        sandwich += ['bread', i, 'bread'] if i == flavor else [i]
    return sandwich
```

```python
if __name__=='__main__':
    print(make_sandwich(["tuna", "ham", "tomato"], "ham"))
```

```python
def percent_filled(box):
	boxSize = 0
	os = 0
	for i in box:
		for j in i:
			if j!= '#':
				boxSize+=1
	for i in box:
		for j in i:
			if j == 'o':
				os+=1
	return str(int(os/boxSize * 100)) + '%'
```

```python
if __name__=='__main__':
    print(percent_filled([
  "####",
  "#  #",
  "#o #",
  "####"
]))
```

```python
def percent_filled(box):
	empty = sum(sum(1 for j in i if j == ' ') for i in box)
	filled = sum(sum(1 for j in i if j == 'o') for i in box)
	return '{}%'.format(round(filled/(filled+empty)*100))
```

```python
if __name__=='__main__':
    print(percent_filled([
  "####",
  "#  #",
  "#o #",
  "####"
]))
```
