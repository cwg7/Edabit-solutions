```python
import math
def eval_factorial(lst):
	newLst = []
	tot = 0
	for i in lst:
		if '!' in i:
			i = i.replace('!','')
			newLst.append(int(i))
	for i in newLst:
		tot += math.factorial(i)
	return tot
```

```python
if __name__=='__main__':
    print(eval_factorial(["5!", "5!", "10!"]))
```

```python
from math import factorial

def eval_factorial(lst):
	return sum(factorial(int(i.replace('!',''))) for i in lst)
```

```python
if __name__=='__main__':
    print(eval_factorial(["5!", "5!", "10!"]))
```
