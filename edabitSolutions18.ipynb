## 2)

```python
import math
def eval_factorial(lst):
	newLst = []
	tot = 0
	for i in lst:
		if '!' in i:
			i = i.replace('!','')
			newLst.append(int(i))
	for i in newLst:
		tot += math.factorial(i)
	return tot
```

```python
if __name__=='__main__':
    print(eval_factorial(["5!", "5!", "10!"]))
```

```python
from math import factorial

def eval_factorial(lst):
	return sum(factorial(int(i.replace('!',''))) for i in lst)
```

```python
if __name__=='__main__':
    print(eval_factorial(["5!", "5!", "10!"]))
```

```python
def is_circle_collision(c1, c2):
    r1, x1, y1 = c1
    r2, x2, y2 = c2
    d = ((x1-x2)**2 + (y1-y2)**2)**0.5
    return d <= r1 + r2
```

```python
if __name__=='__main__':
    print(is_circle_collision([10, 0, 0], [10, 10, 10]))
```

## 4)

```python
def reverse_odd(txt):
	newLst = list(txt.split(' '))
	revWords = []
	for i in newLst:
		if len(i)%2!=0:
			revWords.append(i[::-1])
		else:
			revWords.append(i)
	return ' '.join(revWords)
```

```python
if __name__=='__main__':
    print(reverse_odd("One two three four"))
```

```python
## second solution:
```

```python
def reverse_odd(txt):
	return ' '.join(i[::-1] if len(i)%2 else i for i in txt.split())
```

```python
if __name__=='__main__':
    print(reverse_odd("One two three four"))
```

```python
def rps(p1,p2):
	v = {'Rock':1, 'Paper':2, 'Scissors':3}
	return ["It's a draw", 'The winner is p1', 'The winner is p2'][v[p1]-v[p2]]
```

```python
if __name__=='__main__':
    print(rps("Scissors", "Paper"))
```

## 6)

```python
def intersection_union(lst1, lst2):
	i = list(set(lst1).intersection(set(lst2)))
	s = list(set(lst1).union(set(lst2)))
	return [sorted(i), sorted(s)]
```

```python
if __name__=='__main__':
    print(intersection_union([1, 2, 3, 4, 4], [4, 5, 9]))
```
