## 1)

```python
def word_builder(ltr, pos):
	newWord = ''
	for i in pos:
		newWord += ltr[i]
	return newWord
```

```python
if __name__=='__main__':
    print(word_builder(["g", "e", "o"], [1, 0, 2]))
```

```python
def word_builder(ltr, pos):
	return ''.join(ltr[i] for i in pos)
```

```python
if __name__=='__main__':
    print(word_builder(["g", "e", "o"], [1, 0, 2]))
```

## 2)

```python
def box_seq(step):
	if step == 0:
		return 0
	if step % 2 == 0:
		return step
	return step +2
```

```python
if __name__=='__main__':
    print(box_seq(5))
```

```python
if __name__=='__main__':
    print(box_seq(5))
```

```python
def sum_fractions(lst):
	tot = 0
	index = 0
	for i in lst:
		tot += (lst[index][0]/lst[index][1])
		index +=1
	return round(tot)
```

```python
if __name__=='__main__':
    print(sum_fractions([[18,13], [4,5]]))
```

```python
def sum_fractions(lst):
	return round(sum(n/d for n, d in lst))
```

```python
if __name__=='__main__':
    print(sum_fractions([[18,13], [4,5]]))
```
