## 1)

```python
def word_builder(ltr, pos):
	newWord = ''
	for i in pos:
		newWord += ltr[i]
	return newWord
```

```python
if __name__=='__main__':
    print(word_builder(["g", "e", "o"], [1, 0, 2]))
```

```python
if __name__=='__main__':
    print(word_builder(["g", "e", "o"], [1, 0, 2]))
```

```python
def box_seq(step):
	if step == 0:
		return 0
	if step % 2 == 0:
		return step
	return step +2
```

```python
if __name__=='__main__':
    print(box_seq(5))
```

```python
if __name__=='__main__':
    print(box_seq(5))
```

## 3)

```python
def sum_fractions(lst):
	tot = 0
	index = 0
	for i in lst:
		tot += (lst[index][0]/lst[index][1])
		index +=1
	return round(tot)
```

```python
if __name__=='__main__':
    print(sum_fractions([[18,13], [4,5]]))
```

```python
def sum_fractions(lst):
	return round(sum(n/d for n, d in lst))
```

```python
if __name__=='__main__':
    print(sum_fractions([[18,13], [4,5]]))
```

## 4)

```python
def return_unique(lst):
	newLst = []
	for i in lst:
		if lst.count(i) < 2:
			newLst.append(i)
	return newLst
```

```python
if __name__=='__main__':
    print(return_unique([1, 9, 8, 8, 7, 6, 1, 6]))
```

```python
## second solution:
```

```python
def return_unique(lst):
    return [i for i in lst if lst.count(i)==1]
```

```python
if __name__=='__main__':
    print(return_unique([1, 9, 8, 8, 7, 6, 1, 6]))
```

## 9)

```python
def inclusive_list(start_num, end_num):
	newLst = []
	if start_num > end_num:
		return [start_num]
	for i in range(start_num, end_num+1):
		newLst.append(i)
	return newLst
```

```python
if __name__=='__main__':
    print(inclusive_list(17, 5))
```

```python
## second solution:
```

```python
def inclusive_list(startNum, endNum):
	return list(range(startNum, endNum+1)) or [startNum]
```

```python
if __name__=='__main__':
    print(inclusive_list(17, 5))
```
