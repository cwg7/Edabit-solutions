```python
def is_curzon(num):
	a = 2 ** num + 1
	b = 2 * num + 1
	
	if a % b == 0:
		return True
	else:
		return False
```

### cleaner solution:

```python
def is_curzon(num):
	return (2**num + 1) % (2*num + 1) == 0
```

## 5)

```python
def end_corona(recovers, new_cases, active_cases):
		count = 0
		while active_cases > 0:
			active_cases -= recovers
			active_cases += new_cases
			count += 1
		return count
```

## 6)

```python
def calculator(num1, operator, num2):
	if (operator == '/') and (num2 == 0):
		return "Can't divide by 0!"
	elif operator == '/':
		return num1/num2
	elif operator == '*':
		return num1 * num2
	elif operator == '+':
		return num1 + num2
	elif operator == '-':
		return num1 - num2
```

## 8)

```python
def damage(damage, speed, time):
	if speed < 0 or damage < 0:
		return "invalid"
	if time == "second":
		return damage * speed
	if time == "minute":
		return damage * speed * 60
	if time == "hour":
		return damage * speed * 3600
	
```

## 10)

```python
def jazzify(lst):
    if len(lst) == 0:
        return []
    
    for i in range(len(lst)):
        if lst[i][-1] != '7':
            lst[i] = lst[i] + '7'
    return(lst)
```

```python
if __name__=='__main__':
	print(jazzify(['G','F7','C7']))
```

## 13)

```python
import math

def weight(r, h):
    mass = round(r**2 * h * math.pi,2)
    cmToKgs = .001
    inKgs = round(mass * cmToKgs, 2)
    return inKgs
```

```python
if __name__=='__main__':
	print(weight(5,2))
```

## 15)

```python
def is_triplet(n1, n2, n3):
    lst = [n1,n2,n3]
    sortedLst = sorted(lst)
    n1 = sortedLst[0]
    n2 = sortedLst[1]
    n3 = sortedLst[2]
    
    return ((n1 **2 + n2**2) == n3**2)
    
```

```python
if __name__=='__main__':
	print(is_triplet(4, 3, 5))
```

```python
# cleaner version:

def is_triplet(n1, n2, n3):
	lst = sorted([n1,n2,n3])
	return lst[0]**2 + lst[1]**2 == lst[2]**2
```
