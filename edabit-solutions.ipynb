```python
def is_curzon(num):
	a = 2 ** num + 1
	b = 2 * num + 1
	
	if a % b == 0:
		return True
	else:
		return False
```

### cleaner solution:

```python
def is_curzon(num):
	return (2**num + 1) % (2*num + 1) == 0
```
