## 1)

```python
def is_harshad(num):
	newLst = []
	for i in str(num):
		newLst.append(int(i))
	if num == 0:
		return False
	elif int(num) % sum(newLst) == 0:
		return True
	else:
		return False
```

```python
if __name__=="__main__":
	print(is_harshad(209))
```

```python
def is_harshad(num):
	return num!= 0 and num % sum(int(x) for x in str(num)) == 0
```

```python
if __name__=="__main__":
	print(is_harshad(209))
```

## 2)

```python
def space_weights(planet_a, weight, planet_b):
	planets = {
        "Mercury": 3.7,
        "Venus": 8.87,
        "Earth": 9.81,
        "Mars": 3.711,
        "Jupiter": 24.79,
        "Saturn": 10.44,
        "Uranus": 8.69,
        "Neptune": 11.15
    }
	result = (weight / (planets[planet_a]) * planets[planet_b])
	return round(result, 2)
```

```python
if __name__=="__main__":
	print(space_weights("Earth", 1, "Mars"))
```

## 3)

```python
def remove_smallest(lst):
	if len(lst) == 0:
		return lst
	else:
		lst.remove(min(lst))
		return lst
```

```python
if __name__=="__main__":
	print(remove_smallest([2,2,1,2,1]))
```

## 4)

```python
def cars(wheels, bodies, figures):
	newLst = [wheels//4, bodies//1, figures//2]
	if wheels//4 > 0 and bodies//1 > 0 and figures//2 > 0:
		return min(newLst)
	else:
		return 0
```

```python
if __name__=="__main__":
	print(cars(43, 15, 87))
```

```python
## cleaner solution:
```

```python
def cars(wheels, bodies, figures):
	return min(wheels // 4, bodies, figures // 2)
```

```python
if __name__=="__main__":
	print(cars(43, 15, 87))
```

```python
def prevent_distractions(txt):
    distractions = ('anime', 'meme', 'vine', 'roasts', 'Danny DeVito')
    return 'NO!' if any(i in txt for i in distractions) else 'Safe watching!'
```

```python
if __name__=="__main__":
	print(prevent_distractions('vines that butter my eggroll'))
```

```python
def prevent_distractions(txt):
	lst = ['anime', 'meme', 'vine','roasts', 'Danny DeVito']
	for i in lst:
		if i in txt:
			return 'NO!'
	return 'Safe watching!'
```

```python
if __name__=="__main__":
	print(prevent_distractions('that one actor Danny DeVito'))
```

```python
def equal(a, b, c):
	toStr = str(a) + str(b) + str(c)
	count = 0
	for i in toStr:
		if toStr.count(i) == 3 or toStr.count(i) == 2:
			count = toStr.count(i)
	return count
```

```python
if __name__=="__main__":
	print(equal(6,3,3))
```

## 7)

```python
def can_alternate(s):
	if '1' not in s or '0' not in s:
		return False
	elif s.count('1') - s.count('0') > 1 or s.count('0') - s.count('1') > 1:
		return False
	else:
		return True
```

```python
if __name__=="__main__":
	print(can_alternate('0001111'))
```

```python
## solution 2 using absolute value
```

```python
def can_alternate(s):
    return abs(s.count('0') - s.count('1')) in (0, 1)
```

```python
if __name__=="__main__":
	print(can_alternate('0001111'))
```

```python
def is_vowel_sandwich(s):
	vowels = 'aeiou'
	if len(s) ==3:
		return s[0] not in vowels and s[1] in vowels and s[2] not in vowels
	else:
		return False
```

```python
if __name__=="__main__":
	print(is_vowel_sandwich('bake'))
```

## 11)

```python
def marathon_distance(d):
	newLst = []
	for i in d:
		newLst.append(abs(i))
	return sum(newLst) == 25 if len(d) > 0 else False
```

```python
if __name__=="__main__":
	print(marathon_distance([-6,15,4]))
```

```python
if __name__=="__main__":
	print(marathon_distance([-6,15,4]))
```
