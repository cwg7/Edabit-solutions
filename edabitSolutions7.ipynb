## 1)

```python
def is_harshad(num):
	newLst = []
	for i in str(num):
		newLst.append(int(i))
	if num == 0:
		return False
	elif int(num) % sum(newLst) == 0:
		return True
	else:
		return False
```

```python
if __name__=="__main__":
	print(is_harshad(209))
```

```python
def is_harshad(num):
	return num!= 0 and num % sum(int(x) for x in str(num)) == 0
```

```python
if __name__=="__main__":
	print(is_harshad(209))
```

```python
def space_weights(planet_a, weight, planet_b):
	planets = {
        "Mercury": 3.7,
        "Venus": 8.87,
        "Earth": 9.81,
        "Mars": 3.711,
        "Jupiter": 24.79,
        "Saturn": 10.44,
        "Uranus": 8.69,
        "Neptune": 11.15
    }
	result = (weight / (planets[planet_a]) * planets[planet_b])
	return round(result, 2)
```

```python
if __name__=="__main__":
	print(space_weights("Earth", 1, "Mars"))
```

## 4)

```python
def cars(wheels, bodies, figures):
	newLst = [wheels//4, bodies//1, figures//2]
	if wheels//4 > 0 and bodies//1 > 0 and figures//2 > 0:
		return min(newLst)
	else:
		return 0
```

```python
if __name__=="__main__":
	print(cars(43, 15, 87))
```

```python
def equal(a, b, c):
	toStr = str(a) + str(b) + str(c)
	count = 0
	for i in toStr:
		if toStr.count(i) == 3 or toStr.count(i) == 2:
			count = toStr.count(i)
	return count
```

```python
def can_alternate(s):
	if '1' not in s or '0' not in s:
		return False
	elif s.count('1') - s.count('0') > 1 or s.count('0') - s.count('1') > 1:
		return False
	else:
		return True
```

```python
if __name__=="__main__":
	print(can_alternate('0001111'))
```

```python
## solution 2 using absolute value
```

```python
def can_alternate(s):
    return abs(s.count('0') - s.count('1')) in (0, 1)
```

```python
if __name__=="__main__":
	print(can_alternate('0001111'))
```

## 12)

```python
def ascii_capitalize(txt):
	newStr = ''
	for i in txt:
		if ord(i) % 2 == 0:
			newStr += i.upper()
		else:
			newStr += i.lower()
	return newStr
```

```python
if __name__=="__main__":
	print(ascii_capitalize('to be or not to be!'))
```

```python
## one line solution:
```

```python
def ascii_capitalize(txt):
	return ''.join(i.lower() if ord(i) % 2 else i.upper() for i in txt)
```

```python
if __name__=="__main__":
	print(ascii_capitalize('to be or not to be!'))
```

```python
def trace(lst):
	nums = []
	coord1 = 0
	coord2 = 0
	for i in lst:
		nums.append(lst[coord1][coord2])
		coord1+=1
		coord2+=1
	return sum(nums)
```

```python
if __name__=="__main__":
	print(trace([
                [1, 2, 3],
                [4, 5, 6],
                [7, 8, 9]
]))
```

```python
def convert_cartesian(x, y):
	newLst = []
	index = 0
	for i in x:
		newLst.append([x[index],y[index]])
		index +=1
	return newLst

```

```python
if __name__=="__main__":
	print(convert_cartesian([1, 5, 3, 3, 4], [5, 8, 9, 1, 0]))
```

```python
def split(txt):
    
	def is_vowel(l):
		vowels = 'aeoui'
		return l in vowels
    
	vowelStr = ''
	consStr = ''
	for i in txt:
		if is_vowel(i):
			vowelStr += i
		else:
			consStr += i
	return vowelStr+consStr

```

```python
if __name__=="__main__":
	print(split('abcde'))
```

```python
## solution 2:
```

```python
def split(txt):
	vowels = ''.join(ch for ch in txt if ch in 'aeiouAEIOU')
	others = ''.join(ch for ch in txt if ch not in 'aeiouAEIOU')
	return vowels+others
```

```python
if __name__=="__main__":
	print(split('abcde'))
```
