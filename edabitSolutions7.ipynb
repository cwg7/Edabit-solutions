```python
def is_harshad(num):
	newLst = []
	for i in str(num):
		newLst.append(int(i))
	if num == 0:
		return False
	elif int(num) % sum(newLst) == 0:
		return True
	else:
		return False
```

```python
if __name__=="__main__":
	print(is_harshad(209))
```

```python
def is_harshad(num):
	return num!= 0 and num % sum(int(x) for x in str(num)) == 0
```

```python
if __name__=="__main__":
	print(is_harshad(209))
```
