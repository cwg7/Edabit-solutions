```python
def common_elements(lst1, lst2):
	newLst = []
	for i in lst1:
		for j in lst2:
			if i == j:
				newLst.append(i)
	return sorted(list(set(newLst)))
```

```python
if __name__=='__main__':
    print(common_elements([-1, 3, 4, 6, 7, 9], [1, 3]))
```

```python
def gimme_the_letters(spectrum):
	newLst = list(spectrum.split('-'))
	alphabetlower = 'abcdefghijklmnopqrstuvwxyz'
	alphabethigher = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
	if newLst[0].islower():
		lower = [i for i in alphabetlower if i>= newLst[0] and i<= newLst[1]]
		return ''.join(lower)
	else:
		upper = [i for i in alphabethigher if i>= newLst[0] and i<= newLst[1]]
		return ''.join(upper)
```

```python
if __name__=='__main__':
    print(gimme_the_letters("Q-Z"))
```

```python
if __name__=='__main__':
    print(gimme_the_letters("Q-Z"))
```

```python
def balanced(lst):
	x = len(lst)
	b = int(x/2)
	firstHalf = list(lst[:b])
	secondHalf = list(lst[b:])
	if sum(firstHalf) > sum(secondHalf):
		return firstHalf * 2
	elif sum(secondHalf) > sum(firstHalf):
		return secondHalf * 2
	else:
		return lst
```

```python
if __name__=='__main__':
    print(balanced([7, 5, 2, 6, 1, 0, 1, 5, 2, 7, 0, 6]))
```

```python
def peel_layer_off(lst):
    lst1 = []
    for i in lst[1:-1]:
        lst1.append(i[1:-1])
    return lst1
```

## 6)

```python
def digit_distance(num1, num2):
	str1 = str(num1)
	str2 = str(num2)
	tot = 0
	position = 0
	x = len(str1)
	while x > 0:
		tot += int(str1[position]) - int(str2[position])
		position +=1
		x-=1
	return abs(tot)
```

```python
if __name__=='__main__':
    print(digit_distance(234, 489))
```

```python
## second solution:
```

```python
def digit_distance(num1, num2):
    num1 = [int(i) for i in str(num1)]
    num2 = [int(i) for i in str(num2)]
    return abs(sum([num1[i] - num2[i] for i in range(len(num1))]))
```

```python
if __name__=='__main__':
    print(digit_distance(234, 489))
```

```python
def remove_repeats(s):
	newLst = ['!']
	for i in s:
		if i != newLst[-1]:
			newLst.append(i)
	return ''.join(newLst[1:])
```

```python
if __name__=='__main__':
    print(remove_repeats("aaabbbccc"))
```

```python
def largest_gap(lst):
	x = len(lst)-1
	newLst = sorted(lst)
	pos0=0
	pos1=1
	lst2 = []
	while x > 1:
		lst2.append(newLst[pos1]-newLst[pos0])
		pos0+=1
		pos1+=1
		x-=1 
	return max(lst2)
```

```python
if __name__=='__main__':
    print(largest_gap([9, 4, 26, 26, 0, 0, 5, 20, 6, 25, 5]))
```

```python
## second solution:
```

```python
def largest_gap(lst):
	lst.sort()
	return max(b-a for a, b in zip(lst, lst[1:]))
```

```python
if __name__=='__main__':
    print(largest_gap([9, 4, 26, 26, 0, 0, 5, 20, 6, 25, 5]))
```

```python
import math

def rectangle_in_circle(w, h, radius):
	rectDiag = math.sqrt(w**2 + h**2)
	return rectDiag <= radius*2
```

```python
def validate_subsets(subsets, my_set):
	newLst = []
	for i in subsets:
		for j in i:
			newLst.append(j)
	for i in newLst:
		if i not in my_set:
			return False
	return True
```

```python
if __name__=='__main__':
    print(validate_subsets([[1, 2], [2, 3], [1, 4]], [1, 2, 3]))
```

```python
## second solution:
```

```python
def validate_subsets(subsets, my_set):
	return all(set(x).issubset(my_set) for x in subsets)
```

```python
if __name__=='__main__':
    print(validate_subsets([[1, 2], [2, 3], [1, 4]], [1, 2, 3]))
```
