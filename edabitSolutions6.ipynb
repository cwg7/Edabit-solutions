```python
def nth_smallest(lst, n):
	newLst = sorted(lst)
	if n > len(lst):
		return None
	else:
		return newLst[n-1]
```

```python
if __name__=="__main__":
	print(nth_smallest([7,3,5,1], 2))
```

```python
## alternative solution:
```

```python
def nth_smallest(lst, n):
	return sorted(lst)[n-1] if n <= len(lst) else None
```

```python
if __name__=="__main__":
	print(nth_smallest([7,3,5,1], 2))
```

```python
def is_automorphic(n):
	result = n **2
	newN = str(n)
	newStr = str(result)
	lengthOfN = len(str(n))
	return newN == newStr[-lengthOfN:] 
```

```python
if __name__=="__main__":
	print(is_automorphic(25))
```

```python
## better solution:
```

```python
def is_automorphic(n):
	return str(n**2).endswith(str(n))
```

```python
if __name__=="__main__":
	print(is_automorphic(25))
```

```python
import math

def returnsides(length):
	longest = length * 2
	medium = length * math.sqrt(3)
	return (round(longest, 2), round(medium, 2))
```

```python
def cars_needed(n):

	if n == 0:
		return 0
	elif n <= 5:
		return 1
	elif n % 5 == 0:
		return n//5
	else:
		return n//5 + 1
```

## 8)

```python
def amplify(num):
	lst = []
	for i in range(1, num + 1):
		if i % 4 == 0:
			i = i * 10
			lst.append(i)
		else:
			lst.append(i)
	return lst
```

```python
if __name__=="__main__":
	print(amplify(4))
```

```python
## cleaner solution:
```

```python
def amplify(num):
	return [i*10 if i%4==0 else i for i in range(1,num+1)]
```

```python
if __name__=="__main__":
	print(amplify(4))
```

## 9)

```python
def steps_to_convert(txt):
    uppercase = 0
    lowercase = 0
    for c in txt:
        if c.isupper():
            uppercase += 1
        else:
            lowercase +=1
    if uppercase < lowercase:
        return uppercase
    else:
        return lowercase
```

```python
if __name__=="__main__":
	print(steps_to_convert('abcBBA'))
```

```python
## alternative solution:
```

```python
def steps_to_convert(txt):
	lower = len([i for i in txt if i.islower()])
	upper = len([i for i in txt if i.isupper()])
	return min(lower, upper)
```

```python
if __name__=="__main__":
	print(steps_to_convert('abcBBA'))
```

## 10)

```python
def cap_to_front(s):
	lst1 = []
	lst2 = []
	for i in s:
		if i.islower() == True:
			lst1.append(i)
		else:
			lst2.append(i)
	return ''.join(lst2) + ''.join(lst1)
```

```python
if __name__=="__main__":
	print(cap_to_front('hApPy'))
```

```python
## better solution:
```

```python
def cap_to_front(s):
	uppers = [i for i in s if i.isupper()]
	lowers = [i for i in s if i.islower()]
	return "".join(uppers + lowers)
```

```python
if __name__=="__main__":
	print(cap_to_front('moveMENT'))
```

```python
def get_number_of_apples(n, p):
	percentage = p[:2]
	percentageInt = int(percentage)
	remaining = n - (percentageInt * n / 100)
	if remaining == 0 or n == 0 or p == '100%':
		return "The children didn't get any apples"
	else:
		return int(remaining)
```

```python
if __name__=="__main__":
	print(get_number_of_apples(12, '80%'))
```

```python
if __name__=="__main__":
	print(get_number_of_apples(25, '10%'))
```

```python
def left_digit(num):
	nums = '0123456789'
	for i in num:
		if i in nums:
			return int(i)
```

```python
if __name__=="__main__":
	print(left_digit("TrAdE2W1n95!"))
```

```python
import string

def letters_only(txt):
	alphabet1 = string.ascii_lowercase
	alphabet2 = string.ascii_uppercase
	alphabet3 = alphabet1 + alphabet2
	return ''.join(i for i in txt if i in alphabet3)
```

```python
if __name__=="__main__":
	print(letters_only("R!=:~0o0./c&}9k`60=y"))
```

```python
def letters_only(txt):
	return ''.join(i for i in txt if i.isalpha())
```
