## 1)

```python
def nth_smallest(lst, n):
	newLst = sorted(lst)
	if n > len(lst):
		return None
	else:
		return newLst[n-1]
```

```python
if __name__=="__main__":
	print(nth_smallest([7,3,5,1], 2))
```

```python
## alternative solution:
```

```python
def nth_smallest(lst, n):
	return sorted(lst)[n-1] if n <= len(lst) else None
```

```python
if __name__=="__main__":
	print(nth_smallest([7,3,5,1], 2))
```

## 2)

```python
def maximum_score(tile_hand):
	scores = [i['score'] for i in tile_hand if 'score' in i]
	return sum(scores)
```

```python
if __name__=="__main__":
	print(maximum_score([
        {"tile": "K", "score": 5},
        {"tile": "Z", "score": 10}, 
        {"tile": "X", "score":8},
        {"tile": "D", "score": 2},
        {"tile": "A", "score": 1},
        {"tile": "E", "score": 1}
    ]))
```

```python
## better solution
```

```python
def maximum_score(tile_hand):
	return sum([i['score'] for i in tile_hand])
```

```python
if __name__=="__main__":
	print(maximum_score([
        {"tile": "K", "score": 5},
        {"tile": "Z", "score": 10}, 
        {"tile": "X", "score":8},
        {"tile": "D", "score": 2},
        {"tile": "A", "score": 1},
        {"tile": "E", "score": 1}
    ]))
```

## 3)

```python
def is_automorphic(n):
	result = n **2
	newN = str(n)
	newStr = str(result)
	lengthOfN = len(str(n))
	return newN == newStr[-lengthOfN:] 
```

```python
if __name__=="__main__":
	print(is_automorphic(25))
```

```python
## better solution:
```

```python
def is_automorphic(n):
	return str(n**2).endswith(str(n))
```

```python
if __name__=="__main__":
	print(is_automorphic(25))
```

## 4)

```python
def first_and_last(s):
	return [''.join(sorted(s)), ''.join(sorted(s)[::-1])]
```

```python
if __name__=="__main__":
	print(first_and_last('scoop'))
```
