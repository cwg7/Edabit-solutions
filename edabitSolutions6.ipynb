```python
def nth_smallest(lst, n):
	newLst = sorted(lst)
	if n > len(lst):
		return None
	else:
		return newLst[n-1]
```

```python
if __name__=="__main__":
	print(nth_smallest([7,3,5,1], 2))
```

```python
## alternative solution:
```

```python
def nth_smallest(lst, n):
	return sorted(lst)[n-1] if n <= len(lst) else None
```

```python
if __name__=="__main__":
	print(nth_smallest([7,3,5,1], 2))
```

```python
def maximum_score(tile_hand):
	scores = [i['score'] for i in tile_hand if 'score' in i]
	return sum(scores)
```

```python
if __name__=="__main__":
	print(maximum_score([
        {"tile": "K", "score": 5},
        {"tile": "Z", "score": 10}, 
        {"tile": "X", "score":8},
        {"tile": "D", "score": 2},
        {"tile": "A", "score": 1},
        {"tile": "E", "score": 1}
    ]))
```

## 3)

```python
def is_automorphic(n):
	result = n **2
	newN = str(n)
	newStr = str(result)
	lengthOfN = len(str(n))
	return newN == newStr[-lengthOfN:] 
```

```python
if __name__=="__main__":
	print(is_automorphic(25))
```

```python
## better solution:
```

```python
def is_automorphic(n):
	return str(n**2).endswith(str(n))
```

```python
if __name__=="__main__":
	print(is_automorphic(25))
```

```python
import math

def returnsides(length):
	longest = length * 2
	medium = length * math.sqrt(3)
	return (round(longest, 2), round(medium, 2))
```

```python
if __name__=="__main__":
	print(returnsides(3))
```

## 6)

```python
def cars_needed(n):

	if n == 0:
		return 0
	elif n <= 5:
		return 1
	elif n % 5 == 0:
		return n//5
	else:
		return n//5 + 1
```

```python
if __name__=="__main__":
	print(cars_needed(17))
```

```python
## better solution:
```

```python
from math import ceil

def cars_needed(n):
	return ceil(n/5)
```

```python
if __name__=="__main__":
	print(cars_needed(17))
```

## 8)

```python
def amplify(num):
	lst = []
	for i in range(1, num + 1):
		if i % 4 == 0:
			i = i * 10
			lst.append(i)
		else:
			lst.append(i)
	return lst
```

```python
if __name__=="__main__":
	print(amplify(4))
```

```python
## cleaner solution:
```

```python
def amplify(num):
	return [i*10 if i%4==0 else i for i in range(1,num+1)]
```

```python
if __name__=="__main__":
	print(amplify(4))
```

## 9)

```python
def steps_to_convert(txt):
    uppercase = 0
    lowercase = 0
    for c in txt:
        if c.isupper():
            uppercase += 1
        else:
            lowercase +=1
    if uppercase < lowercase:
        return uppercase
    else:
        return lowercase
```

```python
if __name__=="__main__":
	print(steps_to_convert('abcBBA'))
```

```python
## alternative solution:
```

```python
def steps_to_convert(txt):
	lower = len([i for i in txt if i.islower()])
	upper = len([i for i in txt if i.isupper()])
	return min(lower, upper)
```

```python
if __name__=="__main__":
	print(steps_to_convert('abcBBA'))
```

## 10)

```python
def cap_to_front(s):
	lst1 = []
	lst2 = []
	for i in s:
		if i.islower() == True:
			lst1.append(i)
		else:
			lst2.append(i)
	return ''.join(lst2) + ''.join(lst1)
```

```python
if __name__=="__main__":
	print(cap_to_front('hApPy'))
```

```python
## better solution:
```

```python
def cap_to_front(s):
	uppers = [i for i in s if i.isupper()]
	lowers = [i for i in s if i.islower()]
	return "".join(uppers + lowers)
```

```python
if __name__=="__main__":
	print(cap_to_front('moveMENT'))
```

## 12)

```python
def find_letters(word):
	newLst = []
	for i in word:
		if word.count(i) < 2:
			newLst.append(i)
	return newLst
```

```python
if __name__=="__main__":
	print(find_letters('balloon'))
```

```python
## solution 2:
```

```python
def find_letters(word):
	return [i for i in word if word.count(i) == 1]
```

```python
if __name__=="__main__":
	print(find_letters('balloon'))
```

```python
def get_number_of_apples(n, p):
	percentage = p[:2]
	percentageInt = int(percentage)
	remaining = n - (percentageInt * n / 100)
	if remaining == 0 or n == 0 or p == '100%':
		return "The children didn't get any apples"
	else:
		return int(remaining)
```

```python
if __name__=="__main__":
	print(get_number_of_apples(12, '80%'))
```

```python
if __name__=="__main__":
	print(get_number_of_apples(25, '10%'))
```
