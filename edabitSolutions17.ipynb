```python
def decimal_part(n):
	nStr = str(n)
	lst = list(nStr.split('.'))
	val = int(lst[-1])
	return 0 if len(lst) == 1 else val/(10** len(lst[-1]))
```

```python
if __name__ == '__main__':
    print(decimal_part(-3.73))
```

## 4)

```python
def will_fit(holds, cargo):
	tot = 0
	for i in holds:
		if i == 'M':
			tot +=100
		elif i == 'L':
			tot += 200
		else:
			tot+=0
	return sum(cargo)<tot
```

```python
if __name__ == '__main__':
    print(will_fit(["M", "L", "L", "M"], [56, 62, 84, 90]))
```

```python
def will_fit(holds, cargo):
	d = {'S': 50, 'M': 100, 'L': 200}
	return sum(d[i] for i in holds) >= sum(cargo)
```

```python
if __name__ == '__main__':
    print(will_fit(["M", "L", "L", "M"], [56, 62, 84, 90]))
```

```python
def right_triangle(x, y, z):
	lst = sorted([x,y,z])
	for i in lst:
		if i<=0:
			return False
		else:
			return lst[0]**2 + lst[1]**2 == lst[2]**2
```

```python
if __name__ == '__main__':
    print(right_triangle(3, 4, 5))
```

## 6)

```python
def ends_add_to_10(nums):
	nums = [abs(int(i)) for i in nums]
	newLst = []
	count = 0
	if len(nums) == 0:
		return 0
	for i in nums:
		newLst.append(str(i))
	for j in newLst:
		if int(j[0]) + int(j[-1]) == 10:
			count +=1
	return count
```

```python
if __name__ == '__main__':
    print(ends_add_to_10([33, 44, -55]))
```

## 7)

```python
def is_mini_sudoku(square):
	newLst = []
	for i in square:
		for j in i:
			newLst.append(j)
	return len(set(newLst)) == 9 and 0 not in newLst
```

```python
if __name__ == '__main__':
    print(is_mini_sudoku([[1, 3, 2], [9, 7, 8], [4, 5, 6]]))
```

```python
## second solution:
```

```python
def is_mini_sudoku(square):
	square=[i for x in square for i in x]
	return sorted(square)==[1,2,3,4,5,6,7,8,9]
```

```python
if __name__ == '__main__':
    print(is_mini_sudoku([[1, 3, 2], [9, 7, 8], [4, 5, 6]]))
```

```python
def sum_missing_numbers(lst):
	newLst = []
	for i in range(min(lst), max(lst)+1):
		newLst.append(i)
	return sum(i for i in newLst if i not in lst)
```

```python
if __name__ == '__main__':
    print(sum_missing_numbers([1, 3, 5, 7, 10]))
```

```python
if __name__ == '__main__':
    print(sum_missing_numbers([1, 3, 5, 7, 10]))
```

## 9)

```python
def uncensor(txt, vowels):
	position = 0
	newTxt = ''
	x = len(vowels)
	if x == 0:
		return txt
	while x > 1:
		for i in txt:
			if i == '*':
				i = vowels[position]
				newTxt += i
				position +=1
				x-=1
			else:
				newTxt +=i
				x-=1
	return newTxt
```

```python
if __name__ == '__main__':
    print(uncensor("Wh*r* d*d my v*w*ls g*?", "eeioeo"))
```

```python
## second solution:
```

```python
def uncensor(txt, vowels):
	txt = txt.replace('*', '{}')
	return txt.format(*vowels)
```

```python
if __name__ == '__main__':
    print(uncensor("Wh*r* d*d my v*w*ls g*?", "eeioeo"))
```

```python
def bonus(days):
	x2 = 325
	x3 = 550
	x4 = 600
	tot = 0
	if days < 32:
		return tot
	elif days > 32 and days <= 40:
		tot += (days - 32) * x2
	elif days >= 41 and days <= 48:
		tot += (days - 41) * x3 + (x2 * 8)
	else:
		tot += x2 * 8 + x3 * 8 + (days -48) * x4
	return tot
```

```python
if __name__ == '__main__':
    print(bonus(50))
```

```python
## second solution:
```

```python
def bonus(days):
    total = 0
    for i in range(1, days+1):
        if 32 < i <= 40:
            total += 325
        if 40 < i <= 48:
            total += 550
        if i > 48:
            total += 600
    return total
```

```python
if __name__ == '__main__':
    print(bonus(50))
```

```python
def pluralize(lst):
	newLst = []
	for i in lst:
		if lst.count(i) > 1:
			i = i + 's'
			newLst.append(i)
		else:
			newLst.append(i)
	return set(newLst)
```

```python
if __name__ == '__main__':
    print(pluralize(["cow", "pig", "cow", "cow"]))
```

```python
## second solution:
```

```python
def pluralize(lst):
	return set(i + 's'*(lst.count(i)>1) for i in lst)
```

```python
if __name__ == '__main__':
    print(pluralize(["cow", "pig", "cow", "cow"]))
```

```python
def majority_vote(lst): 
    for i in set(lst):
        if lst.count(i)>len(lst)//2:
            return i
    return None
```

```python
def unstretch(word):
	return word[0] + ''.join(word[i] for i in range(1,len(word)) if word[i] != word[i-1])
```

```python
def str_match_by2char(a, b):
	num = 0
	for i in range(min(len(a),len(b))-1):
		if a[i:i+2] == b[i:i+2]:
			num += 1
	return num
```

```python
if __name__ == '__main__':
	print(str_match_by2char("yytaazz", "yyjaaz"))
```
