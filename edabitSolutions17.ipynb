```python
def circular_shift(lst1, lst2, n):
	return lst2 == lst1[n:]+lst1[:n]
```

```python
if __name__ == '__main__':
    print(circular_shift([1, 2, 3, 4], [3, 4, 1, 2], 2))
```

## 3)

```python
def decimal_part(n):
	nStr = str(n)
	lst = list(nStr.split('.'))
	val = int(lst[-1])
	return 0 if len(lst) == 1 else val/(10** len(lst[-1]))
```

```python
if __name__ == '__main__':
    print(decimal_part(-3.73))
```

## 4)

```python
def will_fit(holds, cargo):
	tot = 0
	for i in holds:
		if i == 'M':
			tot +=100
		elif i == 'L':
			tot += 200
		else:
			tot+=0
	return sum(cargo)<tot
```

```python
if __name__ == '__main__':
    print(will_fit(["M", "L", "L", "M"], [56, 62, 84, 90]))
```

```python
def will_fit(holds, cargo):
	d = {'S': 50, 'M': 100, 'L': 200}
	return sum(d[i] for i in holds) >= sum(cargo)
```

```python
if __name__ == '__main__':
    print(will_fit(["M", "L", "L", "M"], [56, 62, 84, 90]))
```

## 5)

```python
def right_triangle(x, y, z):
	lst = sorted([x,y,z])
	for i in lst:
		if i<=0:
			return False
		else:
			return lst[0]**2 + lst[1]**2 == lst[2]**2
```

```python
if __name__ == '__main__':
    print(right_triangle(3, 4, 5))
```

## 6)

```python
def ends_add_to_10(nums):
	nums = [abs(int(i)) for i in nums]
	newLst = []
	count = 0
	if len(nums) == 0:
		return 0
	for i in nums:
		newLst.append(str(i))
	for j in newLst:
		if int(j[0]) + int(j[-1]) == 10:
			count +=1
	return count
```

```python
if __name__ == '__main__':
    print(ends_add_to_10([33, 44, -55]))
```

## 7)

```python
def is_mini_sudoku(square):
	newLst = []
	for i in square:
		for j in i:
			newLst.append(j)
	return len(set(newLst)) == 9 and 0 not in newLst
```

```python
if __name__ == '__main__':
    print(is_mini_sudoku([[1, 3, 2], [9, 7, 8], [4, 5, 6]]))
```

```python
def is_mini_sudoku(square):
	square=[i for x in square for i in x]
	return sorted(square)==[1,2,3,4,5,6,7,8,9]
```

```python
if __name__ == '__main__':
    print(is_mini_sudoku([[1, 3, 2], [9, 7, 8], [4, 5, 6]]))
```

## 8)

```python
def sum_missing_numbers(lst):
	newLst = []
	for i in range(min(lst), max(lst)+1):
		newLst.append(i)
	return sum(i for i in newLst if i not in lst)
```

```python
if __name__ == '__main__':
    print(sum_missing_numbers([1, 3, 5, 7, 10]))
```

```python
def sum_missing_numbers(lst):
	return sum(range(min(lst),max(lst)+1)) - sum(lst)
```
