```python
def unrepeated(txt):
	newLst = []
	for i in txt:
		if i not in newLst:
			newLst.append(i)
	return ''.join(newLst)
```

```python
if __name__=='__main__':
    print(unrepeated("hello"))
```

```python
def unrepeated(s):
    return ''.join(sorted(set(s), key=s.index))
```

```python
if __name__=='__main__':
    print(unrepeated("hello"))
```

## 2)

```python
def fibonacci(num):
	x = 1000
	fibs = [1,1]
	pos1 = 0
	pos2 = 1
	while x>1:
		fibs.append(fibs[pos1]+fibs[pos2])
		pos1+=1
		pos2+=1
		x-=1
	return fibs[num]
```

```python
if __name__=='__main__':
    print(fibonacci(7))
```

```python
## second solution
```

```python
def fibonacci(num):
	if num < 2:
		return 1
	else:
		return fibonacci(num - 1) + fibonacci(num - 2)
```

```python
if __name__=='__main__':
    print(fibonacci(7))
```

## 3)

```python
def power_of_two(num):
    while num >= 2:
        num /= 2
    return num == 1
```

```python
if __name__=='__main__':
    print(power_of_two(18))
```

## 4)

```python
def num_of_leapyears(years):
	lst = years.split('-')
	lst = [int(i)for i in lst]
	lst2 = []
	for i in range(lst[0], lst[-1]+1):
		if i%4 == 0 and i%100 != 0:
			lst2.append(i)
		elif i%4== 0 and i%100 == 0 and i%400 == 0:
			lst2.append(i)
	return len(lst2)
```

```python
if __name__=='__main__':
    print(num_of_leapyears("1980-1984"))
```

```python
from calendar import isleap

def num_of_leapyears(years):
	start, end = map(int, years.split('-'))
	return sum(isleap(year) for year in range(start, end+1))
```

```python
if __name__=='__main__':
    print(num_of_leapyears("1980-1984"))
```

## 5)

```python
def divisible_by_left(n):
    n = str(n)
    res = [False]
    for i in range(1,len(n)):
        if int(n[i-1]) == 0 or int(n[i]) % int(n[i-1]) != 0:
            res.append(False)
        else:
            res.append(True)
    return res
```

```python
if __name__=='__main__':
    print(divisible_by_left(73312))
```

## 6)

```python
def digits_count(n):
	count = 0
	if n < 0:
		n *= -1
	if n < 10:
		return 1
	while n > 10:
		n/=10
		count+=1
	return count+1
```

```python
if __name__=='__main__':
    print(digits_count(121317))
```
