```python
def unrepeated(txt):
	newLst = []
	for i in txt:
		if i not in newLst:
			newLst.append(i)
	return ''.join(newLst)
```

```python
if __name__=='__main__':
    print(unrepeated("hello"))
```

```python
def unrepeated(s):
    return ''.join(sorted(set(s), key=s.index))
```

```python
if __name__=='__main__':
    print(unrepeated("hello"))
```

## 2)

```python
def fibonacci(num):
	x = 1000
	fibs = [1,1]
	pos1 = 0
	pos2 = 1
	while x>1:
		fibs.append(fibs[pos1]+fibs[pos2])
		pos1+=1
		pos2+=1
		x-=1
	return fibs[num]
```

```python
if __name__=='__main__':
    print(fibonacci(7))
```

```python
def fibonacci(num):
	if num < 2:
		return 1
	else:
		return fibonacci(num - 1) + fibonacci(num - 2)
```

```python
if __name__=='__main__':
    print(fibonacci(7))
```
