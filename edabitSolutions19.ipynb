```python
def unrepeated(txt):
	newLst = []
	for i in txt:
		if i not in newLst:
			newLst.append(i)
	return ''.join(newLst)
```

```python
if __name__=='__main__':
    print(unrepeated("hello"))
```

## 2)

```python
def fibonacci(num):
	x = 1000
	fibs = [1,1]
	pos1 = 0
	pos2 = 1
	while x>1:
		fibs.append(fibs[pos1]+fibs[pos2])
		pos1+=1
		pos2+=1
		x-=1
	return fibs[num]
```

```python
if __name__=='__main__':
    print(fibonacci(7))
```

```python
## second solution
```

```python
def fibonacci(num):
	if num < 2:
		return 1
	else:
		return fibonacci(num - 1) + fibonacci(num - 2)
```

```python
if __name__=='__main__':
    print(fibonacci(7))
```

```python
def num_of_leapyears(years):
	lst = years.split('-')
	lst = [int(i)for i in lst]
	lst2 = []
	for i in range(lst[0], lst[-1]+1):
		if i%4 == 0 and i%100 != 0:
			lst2.append(i)
		elif i%4== 0 and i%100 == 0 and i%400 == 0:
			lst2.append(i)
	return len(lst2)
```

```python
if __name__=='__main__':
    print(num_of_leapyears("1980-1984"))
```

```python
def divisible_by_left(n):
    n = str(n)
    res = [False]
    for i in range(1,len(n)):
        if int(n[i-1]) == 0 or int(n[i]) % int(n[i-1]) != 0:
            res.append(False)
        else:
            res.append(True)
    return res
```

```python
if __name__=='__main__':
    print(divisible_by_left(73312))
```

## 6)

```python
def digits_count(n):
	count = 0
	if n < 0:
		n *= -1
	if n < 10:
		return 1
	while n > 10:
		n/=10
		count+=1
	return count+1
```

```python
if __name__=='__main__':
    print(digits_count(121317))
```

```python
def censor_string(txt, lst, char):
	result= ''
	lst2 = list(txt.split())
	for i in lst2:
		if i in lst:
			i = char * len(i)
			result += i + ' '
		else:
			result += i + ' '
	return result[:-1]
```

```python
def find_broken_keys(txt1, txt2):
    broken_keys = []
    for i in range(len(txt1)):
        if txt1[i] != txt2[i]:
            if txt1[i] not in broken_keys:
                broken_keys.append(txt1[i])
    return broken_keys
```

```python
if __name__=='__main__':
    print(find_broken_keys("happy birthday", "hawwy birthday"))
```

```python
def find_broken_keys(txt1, txt2):
	missing = [a for a, b in zip(txt1, txt2) if a != b]
	return sorted(set(missing), key=txt1.index)
```

```python
if __name__=='__main__':
    print(find_broken_keys("happy birthday", "hawwy birthday"))
```

```python
def vowels(string):
    if len(string) == 0:
        return 0
    if string[0] in 'aeiou':
        return 1 + vowels(string[1:])
    else:
        return vowels(string[1:])
```

```python
if __name__=='__main__':
    print(vowels("a"))
```

## 11)

```python
def perimeter(lst):
	result = 0
	for a,b in lst:
		for x,y in lst:
			result += ((x-a)**2 + (y-b)**2)**0.5
	return round(result/2,2)
```

```python
if __name__=='__main__':
    print(perimeter([[15, 7], [5, 22], [11, 1]]))
```

```python
def is_slidey(num):
    num = str(num)
    for i in range(len(num) - 1):
        if abs(int(num[i]) - int(num[i + 1])) != 1:
            return False
    return True
```

```python
if __name__=='__main__':
    print(is_slidey(123456789))
```

```python
def split_code(item):
	alpha = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ'
	letters = ''
	nums = ''
	for i in item:
		if i in alpha:
			letters += i
		else:
			nums += i
	return [letters,int(nums)]
```

```python
if __name__=='__main__':
    print(split_code("TEWA8392"))
```

```python
## second solution:
```

```python
def split_code(item):
	letters = ''.join(x for x in item if x.isalpha())
	digits = ''.join(x for x in item if x.isnumeric())
	return [letters,int(digits)]
```

```python
if __name__=='__main__':
    print(split_code("TEWA8392"))
```

## 16)

```python
def count_datatypes(*args):
	intCt = 0
	listCt = 0
	strCt = 0
	boolCt = 0
	tupleCt = 0
	dictCt = 0
	for i in args:
		if type(i)==int:
			intCt+=1
		elif type(i)==list:
			listCt+=1
		elif type(i)==str:
			strCt+=1
		elif type(i)==bool:
			boolCt+=1
		elif type(i)==tuple:
			tupleCt+=1
		elif type(i)==dict:
			dictCt+=1
	return [intCt,strCt,boolCt,listCt,tupleCt,dictCt]
```

```python
if __name__=='__main__':
    print(count_datatypes(1, 45, "Hi", False))
```

```python
## much better solution:
```

```python
def count_datatypes(*args):
    lst = [type(i) for i in args]
    return [lst.count(i) for i in (int, str, bool, list, tuple, dict)]
```

```python
if __name__=='__main__':
    print(count_datatypes(1, 45, "Hi", False))
```

```python
def unique_styles(albums):
	lst = []
	lst2 = []
	for i in albums:
		lst.append(i.split(','))
	for i in lst:
		for j in i:
			lst2.append(j)
	return len(set(lst2))
```

```python
if __name__=='__main__':
    print(unique_styles([
  "Dub,Dancehall",
  "Industrial,Heavy Metal",
  "Techno,Dubstep",
  "Synth-pop,Euro-Disco",
  "Industrial,Techno,Minimal"
]))
```

```python
def vowel_links(txt):
    words = txt.split()
    for i in range(len(words) - 1):
        if words[i][-1] in 'aeiou' and words[i + 1][0] in 'aeiou':
            return True
    return False
```

```python
if __name__=='__main__':
    print(vowel_links("a very large appliance"))
```
