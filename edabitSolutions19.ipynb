```python
def unrepeated(txt):
	newLst = []
	for i in txt:
		if i not in newLst:
			newLst.append(i)
	return ''.join(newLst)
```

```python
if __name__=='__main__':
    print(unrepeated("hello"))
```

```python
def unrepeated(s):
    return ''.join(sorted(set(s), key=s.index))
```

```python
if __name__=='__main__':
    print(unrepeated("hello"))
```
