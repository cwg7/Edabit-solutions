## 3)

```python
def reversed_binary_integer(num):
	binNum =  bin(num)[2:]
	revBinNum = str(binNum)[::-1]
	return int(revBinNum,2)
```

```python
if __name__=='__main__':
    print(reversed_binary_integer(10))
```

## 4)

```python
def extend_vowels(word, num):
	vowels = 'aeiouAEIOU'
	result = ''
	if num< 0 or type(num)!= int:
		return 'invalid'
	for i in word:
		if i in vowels:
			result += i * (num+1)
		else:
			result += i
	return result
```

```python
if __name__=='__main__':
    print(extend_vowels("Hello", 5))
```

```python
def extend_vowels(word, num):
	return ''.join((i*num)+i if i in 'aeiouAEIOU' else i for i in word) if num>=0 and int(num)==num else 'invalid'
```

```python
if __name__=='__main__':
    print(extend_vowels("Hello", 5))
```

## 5)

```python
def pentagonal(num):
	x = 1
	dots = 1
	while x < num:
		dots += 5*x
		x+=1
	return dots
```

```python
if __name__=='__main__':
    print(pentagonal(8))
```

```python
## second solution:
```

```python
def pentagonal(num):
	ans = 1
	for i in range(1,num):
		ans += (5*i)
	return ans
```

```python
if __name__=='__main__':
    print(pentagonal(8))
```

## 6)

```python
def encrypt(word):
	codes = {'a':'0','e':'1','i':'2','o':'2','u':'3'}
	revWord = word[::-1]
	result = ''
	for i in revWord:
		if i in codes:
			i = codes[i]
			result += i
		else:
			result += i
	return result + 'aca'
```

```python
if __name__=='__main__':
    print(encrypt("banana"))
```

```python
## second solution:
```

```python
def encrypt(word):
	return word[::-1].translate(str.maketrans('aeou', '0123')) + 'aca'
```

```python
if __name__=='__main__':
    print(encrypt("banana"))
```
