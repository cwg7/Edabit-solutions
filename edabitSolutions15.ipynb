```python
import math
def logarithm(base, num):
    try:
        return math.log(num,base)
    except: return 'Invalid'
```

```python
if __name__=='__main__':
    print(logarithm(5, 25))
```

## 2)

```python
from datetime import date

def is_valid_date(d, m, y):
	try: d = date(y, m, d)
	except: return False
	return True
```

```python
if __name__=='__main__':
    print(is_valid_date(34,2,2020))
```

## 3)

```python
def reversed_binary_integer(num):
	binNum =  bin(num)[2:]
	revBinNum = str(binNum)[::-1]
	return int(revBinNum,2)
```

```python
if __name__=='__main__':
    print(reversed_binary_integer(10))
```
