```python
def factor_chain(lst):
	index1 = 0
	index2 = 1
	iters = len(lst) -1
	while iters > 0:
		if lst[index2] % lst[index1] == 0:
			index1+=1
			index2+=1
			iters-=1
		else:
			return False
	return True
```

```python
if __name__=='__main__':
    print(factor_chain([1,2,4,8,16,32]))
```

## 2)

```python
def length_element(r, i):
	newLst = []
	for j in r:
		newLst.append(j)
	return [len(newLst),newLst[i]]
```

```python
if __name__=='__main__':
    print(length_element(range(40,50,3), 2))
```
