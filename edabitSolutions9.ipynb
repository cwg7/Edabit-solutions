## 1)

```python
def factor_chain(lst):
	index1 = 0
	index2 = 1
	iters = len(lst) -1
	while iters > 0:
		if lst[index2] % lst[index1] == 0:
			index1+=1
			index2+=1
			iters-=1
		else:
			return False
	return True
```

```python
if __name__=='__main__':
    print(factor_chain([1,2,4,8,16,32]))
```

## 2)

```python
def length_element(r, i):
	newLst = []
	for j in r:
		newLst.append(j)
	return [len(newLst),newLst[i]]
```

```python
if __name__=='__main__':
    print(length_element(range(40,50,3), 2))
```

## 3)

```python
def get_frequencies(lst):
	objects = {i : lst.count(i) for i in lst }
	return objects  
```

```python
if __name__=='__main__':
    print(get_frequencies(["A", "B", "A", "A", "A"]))
```

```python
def top_note(student):
	return {"name":student["name"], "top_note": max(student["notes"])}
```

```python
if __name__=='__main__':
    print(top_note({ "name": "John", "notes": [3, 5, 4] }))
```

```python
def add(n1, n2):
	if n1 == '' or n2 == '' or n1 == None or n2 == None:
		return 'Invalid Operation'
	num1 = int(n1)
	num2 = int(n2)
	return str(num1 + num2) 
```

```python
if __name__=='__main__':
    print(add('111', '111'))
```

## 6)

```python
import math
def area_of_hexagon(x):
	return round((3 * math.sqrt(3) * x**2)/2, 1) if x > 0 and type(x) == int else None
```

```python
if __name__=='__main__':
    print(area_of_hexagon(2))
```

```python
def convert_binary(string):
	string = string.lower()
	alphabet = 'abcdefghijklmnopqrstuvwxyz'
	firstHalf = alphabet[:13]
	secondHalf = alphabet[13:]
	result = ''
	for i in string:
		if i in firstHalf:
			result+='0'
		else:
			result +='1'        
	return result
```

```python
if __name__=='__main__':
    print(convert_binary('excLAIM'))
```
