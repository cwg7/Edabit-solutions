```python
def is_symmetrical(num):
	newStr = str(num)
	return newStr[::-1] == newStr
```

```python
if __name__=="__main__":
	print(is_symmetrical(12321))
```

## 2)

```python
def square_digits(n):
	return int("".join(str(int(i)**2) for i in str(n)))
```

```python
if __name__=="__main__":
	print(square_digits(9119))
```

```python
## alternative solution with iteration:
```

```python
def square_digits(n):
    ans = ''
    for i in str(n):
        ans += str(int(i)**2)
    return int(ans)
```

```python
if __name__=="__main__":
	print(square_digits(9119))
```

```python
## by index:
```

```python
def square_digits(n):
	ans = ""
	n = str(n)
	for i in range(len(n)):
		ans += str(int(n[i]) ** 2)
	return int(ans)
```

```python
if __name__=="__main__":
	print(square_digits(2483))
```

```python
def worm_length(worm):
    count = 0
    if len(worm) == 0:
        return 'invalid'
    for i in worm:
        if i != '-':
            return 'invalid'
        else:
            count +=1
    tot = count * 10
    return str(tot) + ' mm.'
```

```python
if __name__=="__main__":
	print(worm_length("----------"))
```
