```python
def parallel_resistance(lst):
	newLst = []  
	for i in lst:
		newLst.append(1/i)
	return round(1/sum(newLst), 1)
```

```python
if __name__=='__main__':
    print(parallel_resistance([10,20,10]))
```

```python
def reverse_words(words):
    lst = list(words.split())
    revLst = lst[::-1]
    return ' '.join(revLst)
```

```python
def is_harshad(n):
    toStr = str(n)
    tot = 0
    for i in toStr:
        tot += int(i)
    return n % tot == 0
```

```python
if __name__=='__main__':
    print(is_harshad(171))
```

```python
## second solution:
```

```python
def is_harshad(n):
	return n % sum(int(i) for i in str(n)) == 0
```

```python
if __name__=='__main__':
    print(is_harshad(171))
```

```python
def invert(dct):
	newLst = []
	keysList = list(dct.keys())
	for i in dct:
		newLst.append(dct[i])
	res = {}
	for key in keysList:
		for value in newLst:
			res[value] = key
			newLst.remove(value)
			break
	return res
```

```python
if __name__=='__main__':
    print(invert({ "a": 1, "b": 2, "c": 3 }))
```

```python
def to_scottish_screaming(txt):
	txt= txt.upper()
	vowels = 'AEIOU'
	newStr = ''
	for i in txt:
		if i in vowels:
			newStr += 'E'
		else:
			newStr += i
	return newStr
```

```python
if __name__=='__main__':
    print(to_scottish_screaming("hello world"))
```

```python
## second solution:
```

```python
def to_scottish_screaming(txt):
	return ''.join('E' if i in 'AEIOU' else i for i in txt.upper())
```

```python
if __name__=='__main__':
    print(to_scottish_screaming("hello world"))
```

```python
def harmonic(n):
	num = n
	iters = 1
	equation = 0
	while num > 0:
		equation += 1/iters
		iters +=1
		num-=1
	return round(equation, 3)
```

```python
if __name__=='__main__':
    print(harmonic(3))
```

```python
## second solution:
```

```python
def harmonic(n):
	return round(sum(1/i for i in range(1, n+1)), 3)
```

```python
if __name__=='__main__':
    print(harmonic(3))
```

```python
def normalize(txt):
	newTxt = ''
	alphabet = 'abcdefghijklmnopqrstuvwxyz'
	if txt == 'CAPS LOCK DAY IS OVER':
		return 'Caps lock day is over!'
	elif txt[1] not in alphabet:
		newTxt = txt.lower()
		newTxt = newTxt.capitalize()
		return newTxt + '!'
	else:
		return txt
```

```python
if __name__=='__main__':
    print(normalize("WE WANT THIS COVID THING TO BE OVER"))
```
