```python
def parallel_resistance(lst):
	newLst = []  
	for i in lst:
		newLst.append(1/i)
	return round(1/sum(newLst), 1)
```

```python
if __name__=='__main__':
    print(parallel_resistance([10,20,10]))
```

```python
def parallel_resistance(lst):
	return round(1 / sum(1/i for i in lst), 1)
```

```python
if __name__=='__main__':
    print(parallel_resistance([10,20,10]))
```

```python
def matrix(a,b,c):
    newLst=[]
    for x in range(a):
        temp_list=[]
        for y in range(b):
            temp_list.append(c)
        newLst.append(temp_list)
    return newLst
```

```python
if __name__=='__main__':
    print(matrix(3, 2, 3))
```

```python
def matrix(x, y, z):
    return [y*[z] for i in range(x)]
```

```python
if __name__=='__main__':
    print(matrix(3, 2, 3))
```

```python
def apocalyptic(n):
	apoc = '666'
	val = 2 ** n
	valStr = str(val)
	if apoc in valStr:
		index = valStr.index(apoc)
		return 'Repent! ' + str(index) + ' days until the Apocalypse!'
	else:
		return 'Crisis averted. Resume sinning.'
```

```python
if __name__=='__main__':
    print(apocalyptic(157))
```

```python
def apocalyptic(n):
    s = str(2**n)
    if '666' in s:
        return "Repent! {} days until the Apocalypse!".format(s.index('666'))
    return "Crisis averted. Resume sinning."
```

```python
if __name__=='__main__':
    print(apocalyptic(157))
```

```python
def tuck_in(lst1, lst2):
	newLst = [lst1[0]]
	for i in lst2:
		newLst.append(i)
	newLst.append(lst1[1])
	return newLst
```

```python
if __name__=='__main__':
    print(tuck_in([1, 10], [2, 3, 4, 5, 6, 7, 8, 9]))
```
