```python
def parallel_resistance(lst):
	newLst = []  
	for i in lst:
		newLst.append(1/i)
	return round(1/sum(newLst), 1)
```

```python
if __name__=='__main__':
    print(parallel_resistance([10,20,10]))
```

```python
def parallel_resistance(lst):
	return round(1 / sum(1/i for i in lst), 1)
```

```python
if __name__=='__main__':
    print(parallel_resistance([10,20,10]))
```

```python
def matrix(a,b,c):
    newLst=[]
    for x in range(a):
        temp_list=[]
        for y in range(b):
            temp_list.append(c)
        newLst.append(temp_list)
    return newLst
```

```python
if __name__=='__main__':
    print(matrix(3, 2, 3))
```

```python
def matrix(x, y, z):
    return [y*[z] for i in range(x)]
```

```python
if __name__=='__main__':
    print(matrix(3, 2, 3))
```
